```python
import random
import matplotlib.pyplot as plt
import statistics as s
```

```python
"Criacao da populacao de 10000 eleitores"

pop = []

pop = [1] * 3000 + [0] * 7000
```

```python
def inferencia_Estatistica(n):

    "Lista que armazena os valores de p chapeu para uma amostra aleatoria"
    p_chapeu = []

    "Amostra aleatoria com n = 360 para 2000 repeticoes"
    for j in range(2000):
        amostra = []

        "Escolha aleatoria de eleitores da populaçao"
        for i in range(n):
            amostra.append(random.choice(pop))

        soma = 0

        "Calculo de p chapeu para cada iteracao"
        for i in range(n):
            soma = soma + amostra[i]
        soma = soma/n

        "Adicao do valor de p chapeu na lista"
        p_chapeu.append(soma)


    "Grafico de valores de p chapeu"
    plt.plot(p_chapeu)
    axes = plt.gca()
    axes.set_ylim([0,1])
    plt.show()

    "Media de valores de p chapeu"
    print("Media: " + str(s.mean(p_chapeu)))

    "Desvio padrao de valores de p chapeu"
    print("Desvio Padrão: " + str(s.stdev(p_chapeu)))
```

```python
inferencia_Estatistica(36)
inferencia_Estatistica(3600)
```

```python
'''A proporção populacional ficou muito próxima da realidade em ambas as amostras, pois em media os valores se aproximam de 0.3. Entretanto, podemos perceber que para o desvio padrão, no caso de uma amostra menor, os valores tendem a se afastar do valor real de p e para o caso de uma amostra grande, o desvio padrao fica menor, pois p chapeu nao se afasta muito do valor real de p'''
```
